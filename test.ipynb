```python

def find_fitness(x_dataset, population, K):
	X=np.array(x_dataset)
	pop=np.array(population)
	pop_distri = []
	for i in pop:
		distri = []
		for j in range(0,K):
			distri.append([])
		for pp in X:
			distance=-1
			ind=-1
			lol=0
			for centroid in i:
				dist = centroid - pp 
				dist=dist*dist
				dist=np.sum(dist)
				if distance < 0:
					distance = dist
					ind = lol
				elif distance > dist: 
					distance = dist 
					ind = lol
				lol = lol + 1
			distri[ind].append(pp)
		pop_distri.append(distri)
	return pop_distri

```
